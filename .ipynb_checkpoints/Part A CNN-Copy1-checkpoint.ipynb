```python
extension_counts = {}

# Function to iterate through folders and count extensions
def count(folder):
    for dirpath, _, filenames in os.walk(folder):
        for filename in filenames:
            _, ext = os.path.splitext(filename)
            if ext:
                ext = ext.lower()
                if ext in extension_counts:
                    extension_counts[ext] += 1
                else:
                    extension_counts[ext] = 1

# Iterate through every folder in CA1 Dataset
for folder_name in os.listdir("C:\\Users\\User\\Downloads\\CA1\\CA1\\Dataset for CA1 part A - AY2425S1"):
    folder_path = os.path.join("C:\\Users\\User\\Downloads\\CA1\\CA1\\Dataset for CA1 part A - AY2425S1", folder_name)
    if os.path.isdir(folder_path):
        count(folder_path)

# Display the count of each extension
for ext, count in extension_counts.items():
    print(f"{ext}: {count}")

# Calculate and display the total number of images
total_images = sum(extension_counts.values())
print("Total Images:", total_images)
```

All file images are in jpg, so we do not need to worry about the file extension

### a iv) Rebalance class weight for 37 by 37 pixel

```python
# CLass weight is imbalance. Use class weight to balance it out
from sklearn.utils.class_weight import compute_class_weight
from sklearn.preprocessing import MultiLabelBinarizer

def generate_class_weights(class_series): 
    class_series = np.argmax(class_series, axis=1)
  
    class_labels = np.unique(class_series)
    class_weights = compute_class_weight(class_weight='balanced', classes=class_labels, y=class_series)
    return dict(zip(class_labels, class_weights))
  
labels = y_train_37
class_weight_37 = generate_class_weights(labels)
```

```python
# Import necessary modules for modelling and evaluation

from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
from tensorflow.keras.layers import Dropout
from tensorflow.keras.layers import Flatten
from tensorflow.keras.layers import Conv2D
from tensorflow.keras.layers import MaxPooling2D
from tensorflow.keras.layers import BatchNormalization

from tensorflow.keras.utils import to_categorical
from tensorflow.keras import regularizers
from keras.optimizers import Adam
from tensorflow.keras.optimizers import SGD
```

# a) 37 pixel by 37 pixel

```python
# Get the number of class for y_test
num_classes = y_train_37.shape[1]
```

```python
# Reshape every X so it can be fit into the model
X_train_37 = X_train_37.reshape(X_train_37.shape[0], 37, 37, 1).astype('float32')
X_test_37 = X_test_37.reshape(X_test_37.shape[0], 37, 37, 1).astype('float32')
X_val_37 = X_val_37.reshape(X_val_37.shape[0], 37, 37, 1).astype('float32')
```

## a i) Base Model 37 by 37 pixel

We create a base model with some layers first and see how it works.

```python
model_37 = Sequential()

model_37.add(Conv2D(64, (5, 5), input_shape=(37, 37, 1), activation='relu'))
model_37.add(Conv2D(16, (3, 3), activation='relu'))
model_37.add(MaxPooling2D(pool_size=(2, 2)))
model_37.add(Flatten())
model_37.add(Dense(64, activation='relu'))

model_37.add(Dense(64, activation='relu'))
model_37.add(Dense(num_classes, activation='softmax'))
```

```python
# Optimizer will be adam

model_37.compile(loss='categorical_crossentropy', 
 optimizer='adam', metrics=['accuracy'])
```

```python
# Show the summary for model 37 pixel
model_37.summary()
```

For accuracy, there seems to be an overfit for model 37 accuracy and loss. To prevent that and to increase accuracy and reduce lose, we will hypertune it by adding layers.

## ii) Hypertune model 37 by 37 pixel (Add some layers) and optimizer

We going to add more layers and use lasso regularization. Not only that, we will add learning rate for the optimizer.

```python
l2_strength = 0.001  # Adjusted for experimentation

model_37_tune = Sequential()

# Adding more convolutional layers with increased filter sizes
model_37_tune.add(Conv2D(32, (5, 5), input_shape=(37, 37, 1), activation='relu'))
model_37_tune.add(BatchNormalization())
model_37_tune.add(MaxPooling2D(pool_size=(2, 2)))

model_37_tune.add(Conv2D(64, (5, 5), activation='relu'))
model_37_tune.add(BatchNormalization())
model_37_tune.add(MaxPooling2D(pool_size=(2, 2)))
model_37_tune.add(Dropout(0.5))

model_37_tune.add(Conv2D(128, (3, 3), activation='relu', kernel_regularizer=regularizers.l2(l2_strength)))
model_37_tune.add(BatchNormalization())
model_37_tune.add(MaxPooling2D(pool_size=(2, 2)))

# Using Global Average Pooling instead of Flatten
model_37_tune.add(Flatten())

# Dense layers with Dropout
model_37_tune.add(Dense(128, activation='relu', kernel_regularizer=regularizers.l2(l2_strength)))
model_37_tune.add(Dropout(0.5))

model_37_tune.add(Dense(64, activation='relu', kernel_regularizer=regularizers.l2(l2_strength)))

model_37_tune.add(Dense(num_classes, activation='softmax'))

# Optimizer will be SGD
optimizer = SGD(learning_rate=0.0001, momentum = 0.8)

model_37_tune.compile(loss='categorical_crossentropy', 
 optimizer=optimizer, metrics=['accuracy'])

# Show the summary for model 37 pixel hypertuned
model_37_tune.summary()
```

```python
# Train model 37 pixel hypertune
history_37_tune = model_37_tune.fit(X_train_37, y_train_37, validation_data=(X_val_37, y_val_37), 
 epochs=50, verbose=1, batch_size = 50, class_weight = class_weights_37) 
```

```python
# Show history of 37 pixel hypertune model for accuracy

plt.plot(history_37_tune.history['accuracy'])
plt.plot(history_37_tune.history['val_accuracy'])
plt.title('Model 37 tune accuracy')
plt.ylabel('accuracy')
plt.xlabel('epoch')
plt.legend(['Accuracy', 'Val_Accuracy'], loc='upper left')
plt.show()
```

```python
# Show history of 37 pixel model for loss
plt.plot(history_37_tune.history['loss'])
plt.plot(history_37_tune.history['val_loss'])
plt.title('Model 37 tune Loss')
plt.ylabel('Validation Loss')
plt.xlabel('epoch')
plt.legend(['loss', 'val_loss'], loc='upper left')
plt.show()
```
